```python
import os
import matplotlib.pyplot as plt

#for loading and visualizing audio files
import librosa
import librosa.display

#to play audio
import IPython.display as ipd

data, rate = librosa.load(faili_nimi, sr=44100)

print("Sampling (frame) rate = ", rate)
print("Total samples (frames) = ", data.shape)
print("Duration in seconds = ", np.count_nonzero(data) / rate)
# print(data)
```

```python
# kolmk√µla amplituud samplite kaupa
plt.figure(figsize=(16, 4))
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.plot(data)
```

```python
# "zoom-in": graafik esimese 20.000 sample'i kohta (u. 0,45 sekundit helifaili algusest)
plt.figure(figsize=(16, 4))
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.plot(data[:20000], '-') # proovi asendada "-" n√§iteks "." v√µi "*"-ga
```

```python
# "zoom-in": graafik esimese 2500 sample'i kohta
plt.figure(figsize=(16, 4))
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.plot(data[:2500], '.');plt.plot(data[:2500], '-');
```

```python
# "zoom-in": graafik sample'ite 500 kuni 700 kohta
plt.figure(figsize=(16, 4))
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.plot(data[500:700], '.');plt.plot(data[500:700], '-');
```

```python
# Fourier teisendus, tulemus omistatakse muutujale x
x = librosa.stft(data)

# Teisendab amplituudi p√µhise spektogrammi dB p√µhiseks
db = librosa.amplitude_to_db(abs(x)) # kasutame amplituudi absoluutv√§√§rtust
```

```python
# Graafiku joonistamine
plt.figure(figsize=(14, 5))
librosa.display.specshow(db, sr=rate, x_axis='time', y_axis='hz')
plt.colorbar() # lisab dB legendi paremale poole
```

```python
# "Zoom-in" - failist loetud kolmk√µla sagedused paistavad olema piirkonnas 0 .. 4000 Hz
plt.figure(figsize=(14, 5))
librosa.display.specshow(db, sr=rate, x_axis='time', y_axis='hz')
plt.ylim([0, 4000]) # piirame graafikul n√§idatavat sagedusvahemikku
plt.colorbar()

# Selgemalt eristuvad 3 sagedust, √ºlej√§√§nu on m√ºra
```

```python
# l√µikame √§ra helil√µigu alguse ja l√µpu: j√§rele j√§√§b l√µik alates 0.25 sek kuni 1.5 sek
alamosa = data[11025:66150] # 11025 = 44100 / 4; 55150 = 44100 * 1.5

# Fourier teisendus, tulemus omistatakse muutujale x
x_alam = librosa.stft(alamosa)

# Teisendab amplituudi p√µhise spektogrammi dB p√µhiseks
db_alam = librosa.amplitude_to_db(abs(x_alam)) # kasutame amplituudi absoluutv√§√§rtust
```

```python
# Uus graafik alaml√µigust
plt.figure(figsize=(14, 5))
librosa.display.specshow(db_alam, sr=rate, x_axis='time', y_axis='hz')
plt.ylim([0, 4000]) # piirame graafikul n√§idatavat sagedusvahemikku
plt.colorbar()
```
