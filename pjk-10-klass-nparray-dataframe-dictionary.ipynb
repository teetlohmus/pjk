```python
import numpy as np # helifaile (wav) käsitletakse Pyhtonis numpy array'dena
import pandas as pd # DataFrame's, csv faili sisselugemine
```

```python
# loome järgmiste näidete uurimiseks df-nimelise dataframe'i:
df = pd.DataFrame({"Veerg1":[14, 4, 5, 4, 1, 11],
                   "Veerg2":[5, 2, 54, 3, 2, 5],
                   "Veerg3":[20, 2092, 7, 3, 8, 20]})

# {} kasutame sõnastiku-tüüpi (ingl. dictionary) objektide korral
# {veeru pealkiri: [väärtused]}

df # kuvab terve dataframe'i (suure andmekogu korral ainult esimesed ja viimased read)
# 0-s veerg on indeks, luuakse automaatselt
```

```python
# [] kasutame DataFrame'i veergudele osutamiseks

print('Valitud veeru väärtused:')
print(df['Veerg1'])

print('-----------------------')

print('Konkreetse lahtri väärtus:')
print(df['Veerg3'][4])
```

```python
# () kasutame funktsioonide kasutamisel

print('Unikaalseid väärtuseid igas veerus:')
print(df.nunique())
```

```python
print('Unikaalseid väärtuseid igas reas:')
print(df.nunique(axis = 1))
```

```python
# Võõrväärtuste (ingl. outlier) olemasolu kontroll:
print(df['Veerg3'].median())
print(df['Veerg3'].mean())
```

```python
# parandame valitud lahtri väärtust
df['Veerg1'][3] = 22
df
# sõltuvalt andmete tähendusest ei saa alati sellist parandust teha 
# nt patsiendi tervisenäitajad - siis peab näidu põhjuseid uurima
```

```python
# Võõrväärtuse (ingl. outlier) uus kontroll:
print(df['Veerg1'].median())
print(df['Veerg1'].mean())
```

```python
# Kõikide veergude keskmine
keskm = df.mean(axis=0)
print(keskm)
```

```python
import pandas as pd

tabel = pd.DataFrame({'Nimi': ['Mirjam','Luukas','Rauno','Mirjam','Luukas','Rauno','Mirjam','Luukas','Rauno','Mirjam','Luukas','Rauno','Mirjam','Luukas','Rauno'],
                      'Kuupäev': ['11.03.2023','11.03.2023','11.03.2023','12.03.2023','12.03.2023','12.03.2023','13.03.2023','13.03.2023','13.03.2023','14.03.2023','14.03.2023','14.03.2023','15.03.2023','15.03.2023','15.03.2023'],
                      'Temp': [36.4, 36.7, 36.6, 36.7, 36.8, 36.6, 37.2, 36.6, 36.4, 37.3, 36.6, 36.5, 36.9, 36.8, 36.6]})
```

```python
print(tabel)
```

```python
# df.loc[df['Fee'] > 22000, 'Fee'] = 15000

# df.loc[tabel['Nimi'] == 'Mirjam']

df.loc[tabel['Nimi'] == 'Mirjam', 'Nimi'] = 'Joosep'
```

```python
tabel['Temp'].mean()
```

```python
tabel_p = pd.pivot_table(tabel, values='Temp', 
                                index='Kuupäev',
                                columns='Nimi')

tabel_p
```

```python
print(tabel_p['Luukas'].mean(), tabel_p['Luukas'].max())
print(tabel_p['Mirjam'].mean())
print(tabel_p['Rauno'].mean())
```

```python
keskm = tabel_p.mean(axis=0)
med = tabel_p.median(axis=0)
maksim = tabel_p.max(axis=0)
minim = tabel_p.min(axis=0)
print (keskm, '\n\n', med, '\n\n', maksim, '\n\n', minim)
```

```python
# Mis on NumPy array? 
# https://www.datacamp.com/tutorial/python-numpy-tutorial
# Seda kasutatakse masinõppes, masinnägemises (computer vision), heli töötluses jpm

p_list = [1, 2, 3, 4]   # Loome listi, paneme sellele nimeks p_list

n_array = np.array(p_list) # Loome 1-D numpy array, paneme nimeks n_array

# funktsioonis kasutame ümarsulge: np.array(...), print(), jne"
# andmeelemendid lisame kandiliste sulgude sees, elementidele viitame kasutades kandilisi sulge
```

```python
print ('list: ', p_list)
print ('np_array: ', n_array)

# tunduvad sarnased ?

# mis muutub nii listis kui array's, kui eelmises plokis teha selline muudatus: p_list = [1, 2, '3', 4] ?
```

```python
# list ja numpy array käituvad väga erinevalt:

print('List + List = ', p_list + p_list)
print('Array + Array = ', n_array + n_array)

print('List x 3 = ', p_list * 3)
print('Array x 3 = ', n_array * 3)
```
