```python
import numpy as np # helifaile (wav) käsitletakse Pyhtonis numpy array'dena
import pandas as pd # DataFrame's, csv faili sisselugemine
```

```python
# Mis on NumPy array? 
# https://www.datacamp.com/tutorial/python-numpy-tutorial
# Seda kasutatakse masinõppes, masinnägemises (computer vision), heli töötluses jpm

p_list = [1, 2, 3, 4]   # Loome listi, paneme sellele nimeks p_list

n_array = np.array(p_list) # Loome 1-D numpy array, paneme nimeks n_array

# funktsioonis kasutame ümarsulge: np.array(...), print(), jne"
# andmeelemendid lisame kandiliste sulgude sees, elementidele viitame kasutades kandilisi sulge
```

```python
print ('list: ', p_list)
print ('np_array: ', n_array)

# tunduvad sarnased ?

# mis muutub nii listis kui array's, kui eelmises plokis teha selline muudatus: p_list = [1, 2, '3', 4] ?
```

```python
# list ja numpy array käituvad väga erinevalt:

print('List + List = ', p_list + p_list)
print('Array + Array = ', n_array + n_array)

print('List x 3 = ', p_list * 3)
print('Array x 3 = ', n_array * 3)
```
