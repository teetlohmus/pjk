```python
import matplotlib.pyplot as plt
figure, axes = plt.subplots(facecolor='#FFEFDE') # loome joonise ja teljed ning defineerime ümbrisvärvi 
raadius_1 = 0.3 # defineerime muutuja: 1. ringi raadius
raadius_2 = raadius_1 - 0.1 # defineerime muutuja: 2. ringi raadius
raadius_3 = raadius_2 - 0.1 # defineerime muutuja: 3. ringi raadius
hor_keskpunkt = 0.5 # defineerime muutuja: ringide keskpunkt x-teljel
vert_keskpunkt = 0.4 # defineerime muutuja: ringide keskpunkt y-teljel
ring_1 = plt.Circle(( hor_keskpunkt , vert_keskpunkt ), raadius_1, color='#0000FF' ) # 1.ringi parameetrid
ring_2 = plt.Circle(( hor_keskpunkt , vert_keskpunkt ), raadius_2, color='#00FF00' ) # 2.ringi parameetrid
ring_3 = plt.Circle(( hor_keskpunkt , vert_keskpunkt ), raadius_3, color='#FF0000' ) # 3.ringi parameetrid
axes.set_aspect('equal') # seadistame x- ja y-telje omavahelise suhte (aspect ratio)
axes.add_artist( ring_1 ) # lisame 1. ringi teljestikule
axes.add_artist( ring_2 ) # lisame 2. ringi teljestikule
axes.add_artist( ring_3 ) # lisame 3. ringi teljestikule
axes.set_facecolor("#76C268") # seadistame taustavärvi
pealkiri = 'Ringid keskpunktiga x = ' + str(hor_keskpunkt) + ' ja y = ' + str(vert_keskpunkt) # defineerime muutuja: joonise pealkiri
plt.title(pealkiri, color='#38761D') # seadistame joonise pealkirja ja selle värvi"
plt.show() # näita joonist
# laiem värvide valik: https://www.color-hex.com/ või Google
```

```python
import matplotlib.pyplot as plt

varvid = ['#FF0000','#00FF00','#0000FF']
raadiused = [0.3, 0.2, 0.1]

joonis, teljed = plt.subplots()

teljed.set_aspect('equal')

for i in range(3):
    teljed.add_artist(plt.Circle((0.5 , 0.5), raadiused[i], color=varvid[i]))

plt.show()
```
